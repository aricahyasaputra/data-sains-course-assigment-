# Latihan (12)

Visualisasi Boxplot untuk memberikan ringkasan statis dari fitur yang diplot. 
- Garis atas mewakili nilai maksimal

# Latihan (18)

Visualisasi Stacked Histogram digunakan untuk menunjukkan bagaimana fitur  yang lebih besar dibagi menjadi fitur yang lebih kecil dan menunjukan hubungan masing-masing fitur terhadap jumlah total

Tampilkan visualisasi Stacked Histogram dari fitur "Entity" dengan mengubah tipe datanya menjadi *category* (astype) dalam dataset death rates from air pollution

Agar data dapat diolah maka data harus di rename terlebih dahulu nama kolomnya

```python
df_rename = df.rename(columns = {'Air pollution (total) (deaths per 100,000)':'AirPollution'})
df_rename.head()
```

```python
# visualisasi stacked histogram
df_rename['Entity'] = df_rename['Entity'].astype('category')
#df.head()
list1=list()
mylabels=list()
for gen in df_rename.Entity.cat.categories:
    list1.append(df_rename[df_rename.Entity==gen].AirPollution)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=20,stacked=True,rwidth=1,label=mylabels,range = (0,20))
plt.legend(frameon=False)
plt.show()
```
