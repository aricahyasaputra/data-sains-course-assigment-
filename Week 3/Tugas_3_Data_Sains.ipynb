#Mencari Pencilan menggunakan metode Turkey Fences

```python
q1 = df_noid.quantile(0.25)
q3 = df_noid.quantile(0.75)
iqr = q3 - q1
iqr
```

```python
import warnings
warnings.filterwarnings('ignore')

df_noid_align, iqr_new = df_noid.align(iqr, axis=1, copy=False, join='outer')
outlier_filter = (df_noid < q1 - 1.5 * iqr_new) | (df_noid > q3 + 1.5 * iqr_new)
outlier_filter
```

Karena tidak terdapat data nama maka saya bandingkan berdasarkan kills dan gold

```python
df_noid [outlier_filter['kills']]\
  .loc[:, ['kills', 'gold']]\
  .sort_values(by=['kills'], ascending=False)
```
